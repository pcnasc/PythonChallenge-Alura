Desafio:

Criar um arquivo .txt a partir de uma lista de perguntas vindas de uma lista em Python [].
Ler as perguntas desse arquivo .txt e salvá-las em uma lista Python
Obter respostas de um LLM para cada uma.
Salvar os resultados em um novo arquivo .csv
Ler o arquivo .csv usando Pandas

```python
from google import genai
client = genai.Client(api_key="xxxx")
```

```python
# Etapa 1
questions_list = [
    "What is the compression ratio of the Bugatti Chiron's 8.0L quad-turbocharged W16 engine?",
    "What type of valve train system does the Honda K20 engine use?",
    "What is the difference between understeer and oversteer, and how does weight distribution affect each?",
    "How does a Limited-Slip Differential (LSD) differ from an open differential in torque distribution?",
    "What is the purpose of a harmonic balancer on a crankshaft, and what happens when it fails?"
]

with open("questions.txt", "w", encoding="utf-8") as file:
    for question in questions_list:
        file.write(question + "\n")
```

```python
# Etapa 2
challenge_list = []

with open("questions.txt", "r", encoding="utf-8") as file:
    for line in file:
        challenge_list.append(line.strip())
    print(challenge_list)
```

```python
import time

list_of_answers = []

for questions in challenge_list:
    while True:
        try:
            answer = client.models.generate_content(
                model="gemini-2.5-flash",
                contents=f"generate a succinct answer for the question {questions}"
            )
            list_of_answers.append({"question": questions, "answer": answer.text})
            break
        except Exception as e:
            print(f"Error: {e}. Retrying in 5 seconds...")
            time.sleep(5)
```

```python
list_of_answers
```

```python
import pandas as pd
df_questions_and_answers = pd.DataFrame(list_of_answers)
```

```python
df_questions_and_answers
```

```python
df_questions_and_answers.to_csv("answers.csv", index=False, encoding="utf-8")
```
